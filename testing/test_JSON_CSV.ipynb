```python
import datetime

def _format_time(date):
    #
    #2020-03-27 09:32
    d_f = date.count('-')
    t_f = date.count(':')
    if d_f == 2:
        date_format = "%Y-%m-%d"
    elif d_f == 1:
        date_format = "%m-%d"
    else:
        date_format = ''

    if t_f == 2:
        time_format = '%H:%M:%S'
    elif t_f == 1:
        time_format = '%H:%M'
    else:
        time_format = ''
    format_string = " ".join([date_format, time_format]).rstrip()
    return (format_string)
    
_format_time("2020-03-27 09:32")
    
    
```
